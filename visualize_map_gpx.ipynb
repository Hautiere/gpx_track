```python
import folium
import gpxpy

# Charge le fichier GPX
with open('parcours.gpx', 'r') as f:
    gpx = gpxpy.parse(f)

# Crée une liste de points à partir des points de tracé du fichier GPX
points = []
for track in gpx.tracks:
    for segment in track.segments:
        for point in segment.points:
            points.append(tuple([point.latitude, point.longitude]))

# Crée une carte centrée sur le parcours
m = folium.Map(location=points[0], zoom_start=13)

# Ajoute le parcours à la carte en utilisant une ligne en forme de "trail"
folium.PolyLine(points, color="red", weight=2.5, opacity=1,
                trail_color="#6b6b6b").add_to(m)
```

```python
import gpxpy
import folium

# Read in the GPX file
with open('parcours.gpx', 'r') as f:
    gpx = gpxpy.parse(f)

# Create a map centered at the starting point of the route
start_lat, start_lon = gpx.tracks[0].segments[0].points[0].latitude, gpx.tracks[0].segments[0].points[0].longitude
map = folium.Map(location=[start_lat, start_lon], zoom_start=12)

# Add the route to the map
folium.PolyLine(locations=[[p.latitude, p.longitude] for p in gpx.tracks[0].segments[0].points], color='red').add_to(map)

# Display the map
map
```
